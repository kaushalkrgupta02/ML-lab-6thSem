```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X_iris, y_iris = iris.data[:, :2] , iris.target
```

```python
X_iris_train, X_iris_test, y_iris_train, y_iris_test = train_test_split(X_iris,y_iris,test_size=0.2,random_state=42)
```

```python
svm_iris = SVC(kernel='linear')
```

```python
svm_iris.fit(X_iris_train,y_iris_train)
```

```python
y_iris_pred = svm_iris.predict(X_iris_test)
```

```python
print("Accuraccy of SVM on iris :", accuracy_score(y_iris_test,y_iris_pred))
```

```python
print("MSE of SVM on iris :", mean_squared_error(y_iris_test,y_iris_pred))
```

```python
plt.figure(figsize=(6,4))
plt.scatter(range(len(y_iris_test)),y_iris_test,label='Actual' ,color='black',marker='o')
plt.scatter(range(len(y_iris_pred)),y_iris_pred,label='Predicted' ,color='red',marker='x')
plt.title("Iris Dataset - Actual vs Predicted Labels (SVM)")
plt.xlabel("Sample Index")
plt.ylabel("Class Label")
plt.legend()
plt.grid(True)
plt.show()
```
