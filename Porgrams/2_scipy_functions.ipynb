```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize, integrate, stats
```

```python
#Minimize the function f(x) = (x - 3)^2 + 2

def func(x):
    return (x - 3)**2 + 2

res = optimize.minimize(func, x0=0)
print(f"Minimum at x = {res.x[0]:.2f}, f(x) = {res.fun:.2f}")
```

```python
# Working with Normal Distribution and Probability Density Function (PDF)

# PDF and random sampling from standard normal
x = np.linspace(-3, 3, 100)
pdf = stats.norm.pdf(x)
samples = stats.norm.rvs(size=1000)

plt.plot(x, pdf, label='PDF')
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Samples')
plt.legend()
plt.title('Normal Distribution')
plt.show()
```
